```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Practice').getOrCreate()
```

```python
spark
```

```python
# Loading dataset to py_spark dataframe
df_pyspark = spark.read.csv('/content/drive/MyDrive/Datasets/Pyspark/test1.csv', header=True, inferSchema=True)
df_pyspark.show()
```

## Filter operations

```python
# selecting specific columns with salary < 20000
# method 1
df_pyspark.filter('Salary<20000').select(['Name', 'age']).show()
```

```python
# Method 2
df_pyspark.filter(df_pyspark['Salary'] < 20000).select(['Name', 'age']).show()
```

```python
# Multiple filter conditions
df_pyspark.filter((df_pyspark['Salary'] < 20000) &
                  (df_pyspark['age'] > 21)).show()
```

```python
# Negation operation
df_pyspark.filter(~(df_pyspark['Salary'] < 20000)).show()
```
